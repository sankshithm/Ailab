```python
cars_df=pd.read_csv("C:\\Users\\GPT BANTWAL\\Downloads\\Cars931611042729436\\Cars93.csv")
cars_df
```

```python
df=cars_df.dropna(axis=0)
df.isnull().sum()
```

```python
df=cars_df.drop(['Model'],axis=1)
df.isnull().sum()
```

```python
cars_df['Model']=cars_df['Model'].fillna(0)
cars_df['Model'].isnull().sum()
```

```python
cars_df['Length'] = cars_df['Length'].fillna(cars_df['Length'].mean())
cars_df['Length']
```

```python
cars_df['Luggage.room'] = cars_df['Luggage.room'].fillna(cars_df['Luggage.room'].mode()[0])
cars_df['Luggage.room']
```

```python
cars_df['MPG.highway']=cars_df['MPG.highway'].fillna(cars_df['MPG.highway'].median())
cars_df['MPG.highway']
```

```python
 credit_df=cars_df.filna(method='ffill')
 credit_df   
```
