# Imputing the missing value

# Replacing with arbitrary value

```python
titanic_df['deck'].unique()
```

```python
titanic_df['deck']=titanic_df['deck'].fillna('c')
```

```python
titanic_df['deck'].isnull().sum()
```

```python
titanic_df
```

# Replacing with Mean

```python
mean=titanic_df['age'].mean()
print(mean)
titanic_df['age']=titanic_df['age'].fillna(mean)
titanic_df['age']
```

# Using missingno to visualize missing value

```python
msno.bar(titanic_df)
```

```python
msno.matrix(titanic_df)
```

# Replacing with Median

```python
titanic_df['age']=titanic_df['age'].fillna(titanic_df['age'].median())
titanic_df['age']
```

# Forword and backword filling if missing values

```python
titanic_df = pd.read_csv("C:\\Users\\GPT BANTWAL\\Downloads\\titanic.csv")
titanic_df
```

```python
new_df = titanic_df.fillna(method="ffill")
new_df
```

```python
new_df = titanic_df.fillna(method="ffill",limit=1)
new_df
```
